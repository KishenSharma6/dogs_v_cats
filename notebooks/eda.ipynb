```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

**Image EDA - Cats**

```python
from turtle import left
from PIL import Image
from random import randint

path = "/home/kishen/documents/python_projects/dogs_v_cats/data/train/"

#Randomly select 10 images to display
rand_int = [randint(0, 500) for p in range(0, 10)]

#Build plot
f, ax = plt.subplots(1, 10, figsize = (25, 10))

for i in range(len(rand_int)):
    im = Image.open(path + "cat." + str(i) + ".jpg")
    ax[i].imshow(im)
    ax[i].axis('off')

f.suptitle("Cat Images", y=.6)
```
