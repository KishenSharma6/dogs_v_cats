```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

**Image EDA - Cats**

```python
from turtle import left
from PIL import Image
from random import randint

path = "/home/kishen/documents/python_projects/dogs_v_cats/data/train/"

#Randomly select 10 images to display
rand_int = random.sample(range(0, 12_500), 10)

#Build plot
f, ax = plt.subplots(1, 10, figsize = (25, 10))

for i in range(len(rand_int)):
    im = Image.open(path + str(rand_int[i]) + ".jpg")
    w, h = im.size
    ax[i].imshow(im)
    ax[i].axis('off')
    ax[i].set_title("Shape: "+ str(w)+ ',' +str(h))

f.suptitle("Cat Images", y=.67)
```
